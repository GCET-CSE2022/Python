```python
a = 10
print("Integer:", a)
print("Type:", type(a))
print("ID:", id(a))  # Memory address
b = 3.14
print(f"\nFloat: {b}, Type: {type(b)}, ID: {id(b)}")
c = 2 + 3j
print(f"\nComplex: {c}, Type: {type(c)}, ID: {id(c)}")
d = True
print(f"\nBoolean: {d}, Type: {type(d)}, ID: {id(d)}")
e = "Faheem was here"
print(f"\nString: {e}, Type: {type(e)}, ID: {id(e)}")
r = range(1, 6)
print(f"\nRange: {list(r)}, Type: {type(r)}")


a = 10
print(float(a))  # Output: 10.0
b = 5.75
print(int(b))  # Output: 5
c = "123"
print(int(c))  # Output: 123
complex_number = complex(a)
print(complex_number)  # Output: (10+0j)
flag = True
print(int(flag))  # Output: 1
num = 25
print(f"Decimal : {num}")
print(f"Decimal to Binary : {bin(num)}")  # Output: 0b11001
print(f"Decimal to Octal : {oct(num)}")   # Output: 0o31
print(f"Decimal to Hexadecimal : {hex(num)}")  # Output: 0x19

v_binary = "11001"
b_to_dec = int(v_binary, 2)  # Convert binary to decimal
print(f"Binary : {v_binary}")
print(f"Binary to Decimal : {b_to_dec}")
print(f"Binary to Octal : {oct(b_to_dec)}")
print(f"Binary to Hex : {hex(b_to_dec)}")

octal_val = "34"
oct_to_dec = int(octal_val, 8)  # Convert octal to decimal
print(f"Octal Value : {octal_val}")
print(f"Octal to Decimal : {oct_to_dec}")
print(f"Octal to Binary : {bin(oct_to_dec)}")
print(f"Octal to Hex : {hex(oct_to_dec)}")

hex_val = "19"
hex_to_dec = int(hex_val, 16)  # Convert hex to decimal
print(f"Hex Value : {hex_val}")
print(f"Hex to Decimal : {hex_to_dec}")
print(f"Hex to Octal : {oct(hex_to_dec)}")
print(f"Hex to Binary : {bin(hex_to_dec)}")

```
